```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("expenseImage.png")
```

```python
output = reader.readtext("expenseImage.png")
```

```python
output
```

```python
exp_vals =[]
exp_labels =[]

print(len(output))
index = 0

while index<len(output):
  if index % 2 == 0 :
    exp_labels.append(str(output[index][1]))
    index += 1
  else:
    exp_vals.append(int(output[index][1]))
    index += 1
print(exp_vals)
print(exp_labels)
  
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.2f%%',explode=[0,0,1,0,0],shadow=True)
plt.show()
```
